```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Generate fake sales data (100 rows)
np.random.seed(42)  # For reproducibility
dates = pd.date_range(start='2023-01-01', end='2023-12-31', freq='D')
products = ['Laptop', 'Phone', 'Tablet', 'Headphones', 'Mouse']
regions = ['North', 'South', 'East', 'West']

data = {
    'Date': np.random.choice(dates, 100),
    'Product': np.random.choice(products, 100),
    'Quantity': np.random.randint(1, 10, 100),
    'Price': np.random.uniform(50, 500, 100),
    'Region': np.random.choice(regions, 100)
}
df = pd.DataFrame(data)
df['Total Sales'] = df['Quantity'] * df['Price']  # Add calculated column

# If using a real CSV: df = pd.read_csv('your_file.csv')
print(df.head())  # Preview first 5 rows
```
